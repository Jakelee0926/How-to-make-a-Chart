```python
import matplotlib.pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("GDP per capital")
plt.ylabel("dollars")
plt.savefig("gdp_per_capita.png", dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = []
y = []
for i in range(-10, 10):
  x.append(i)
for t in x:
  y.append(t*2)
plt.plot(x, y, color='blue', marker='o', linestyle='solid')
plt.title("y = 2x")
plt.ylabel("y")
plt.xlabel("x")
plt.axis([-20, 20, -20, 20])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = []
y = np.random.uniform(-3, 3, 1000)
for i in range(0, 1000):
  x.append(i)
plt.plot(x, y, color='red', marker='o', linestyle='solid')
plt.title("numbers")
plt.ylabel("y")
plt.xlabel("x")
plt.show()
```

```python
x = [x for x in range(-20, 20)]
y1 = [t*2 for t in x]
y2 = [t**2 + 5 for t in x]
y3 = [-t**2 - 5 for t in x]
plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.axis([-30, 30, -30, 30])
plt.show()
```

```python
x = [(x/10) for x in range(20)]
y = [(x/10)**2 for x in range(20)]
z = [(x/10)**3 for x in range(20)]
a = [2**(x/10) for x in range(20)]
plt.plot(x, x, label='linear')
plt.plot(x, y, label='quadratic')
plt.plot(x, z, label='cubic')
plt.plot(x, a, label='power')
plt.title("My plot")
plt.ylabel('y label')
plt.xlabel('x label')
plt.legend()
plt.axis([0, 2, 0, 8])
plt.show()
```

```python
import math
x = [x for x in range(0,360)]
y = [math.sin(math.radians(t)) for t in x]
z = [math.cos(math.radians(t)) for t in x]
plt.plot(x, z, color='red', label='cos')
plt.plot(x, y, color='blue', label='sin')
plt.ylabel('y label')
plt.xlabel('x label')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.bar(range(len(years)), gdp)

plt.title("GDP per capital")
plt.ylabel("dollars")
plt.xticks(range(len(years)), years)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xData = np.arange(20, 50)
yData = xData + 2*np.random.randn(30)

plt.scatter(xData, yData)
plt.title('Real Age ve Physical Age')
plt.xlabel('Real Age')
plt.ylabel('Physical Age')
plt.show()
```
