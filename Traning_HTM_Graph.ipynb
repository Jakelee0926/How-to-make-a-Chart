```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)
temp = []
averages = 0
for row in data:
  if row[2] != '':
    year, mon, day = map(int, row[0].split('-'))
    if year >= 1907 and year <= 2019 and mon == 8:
      temp.append(float(row[2]))
      averages += float(row[2])
print(averages/len(temp))
plt.plot(temp, 'r')
plt.show
```

```python
import csv

f=open('seoul.csv', encoding='cp949')
data=csv.reader(f)
next(data)
temp = []
cold = []
for row in data:
  if row[-1] != '' and row[-2] != '':
    year, mon, day = map(int, row[0].split('-'))
    if year >= 1907 and year <= 2019:
      temp.append(float(row[-1]))
      cold.append(float(row[-2]))
print(max(temp))
print(min(cold))
```
